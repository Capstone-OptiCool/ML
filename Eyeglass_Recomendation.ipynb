```python
df = pd.read_csv('cleaned_data_final.csv')

data_frame = df
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df.head()
```

```python
#gabungkan kolom menjadi satu teks untuk rekomendasi
df['combined_column'] = df['Name'] + ' ' + df['Brand'] + ' ' + df['FaceShape'] + ' ' + ' ' + df['Gender'] + ' ' +df['FrameColour'] + ' ' +df['FrameStyle'] + ' ' +df['FrameMaterial']
```

```python
df['combined_column'].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['combined_column'].fillna(''))

# Cek shape matrix hasil TF-IDF
print(tfidf_matrix.shape)
```

```python
tfidf_matrix
```

```python
tfidf_matrix.toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Hitung similarity score antar kacamata
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Buat dictionary untuk mapping antara judul dengan indeks data
indices = pd.Series(df.index, index=df['Name']).drop_duplicates()


# Fungsi untuk mendapatkan rekomendasi berdasarkan similarity score
def get_recommendations(name, cosine_sim=cosine_sim, indices=indices):
    idx = indices[name]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:16]  # Ambil 15 dokumen paling mirip
    movie_indices = [i[0] for i in sim_scores]
    return df['Name'].iloc[movie_indices]
```

```python
recommendations = get_recommendations('Oakley OX8156 HOLBROOK RX 815601')
print(recommendations)
```
